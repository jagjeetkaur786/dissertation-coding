```python
# Function to preprocess datasets
def preprocess_data(df, relevant_columns):
    # Drop unnecessary columns
    df = df[relevant_columns]

    # Drop rows with null values
    df.dropna(inplace=True)

    # Convert 'dt' to datetime format
    df['dt'] = pd.to_datetime(df['dt'])

    return df

# Define relevant columns for each dataset
global_columns = ['dt', 'LandAverageTemperature', 'LandAndOceanAverageTemperature']
city_columns = ['dt', 'City', 'Country', 'AverageTemperature']
country_columns = ['dt', 'Country', 'AverageTemperature']
state_columns = ['dt', 'Country', 'State', 'AverageTemperature']
major_city_columns = ['dt', 'City', 'Country', 'AverageTemperature']

# Preprocess each dataset
global_temps = preprocess_data(global_temps, global_columns)
city_temps = preprocess_data(city_temps, city_columns)
country_temps = preprocess_data(country_temps, country_columns)
state_temps = preprocess_data(state_temps, state_columns)
major_city_temps = preprocess_data(major_city_temps, major_city_columns)
```

```python
# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Decision Tree Regression': DecisionTreeRegressor()
}

# Store evaluation metrics
results = {}

# Train models and evaluate performance
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results[name] = {'MSE': mse, 'R²': r2}

# Print evaluation metrics
for model_name, metrics in results.items():
    print(f"{model_name} - MSE: {metrics['MSE']}, R²: {metrics['R²']}")
```
